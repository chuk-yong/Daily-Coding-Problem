## Find Intersection of linked lists
Given two singly linked lists that intersect at some point, find the intersecting node.  Assuming the nodes are non-cyclical.

For example 
```
A = 3->7->8->10
B = 99->1->8->10
```
return the node with value 8.

Do this in $O(m+n)$ time where m and n are the length of the lists and constant space.

###brute force/naive method
put the shorter list into a set, then iterate over the other to see if any of the node is in the set. But this would take  ùëÇ(ùëö+ùëõ)  time and  ùëÇ(ùëö+ùëõ)  space.

### A better explanation

We have 2 lists that somehow got merged together at some point.  Find that node.

With the diagram, the solution makes sense. 

We find the length of the 2 nodes. Take the difference and start the longer one at the difference.  Move the 2 pointers together.

image.png

````python
## solution from Book
def length(ll): # recursive solution
  if not ll:
    return 0
  return 1+length(ll.next)

def intersect(ll_a,ll_b):
  m, n = length(ll_a), length(ll_b)
  cur_ll_a, cur_ll_b = ll_a, ll_b

  if m > n:
    for _ in range(m-n):
      cur_ll_a = cur_ll_a.next
    else:
      cur_ll_b = cur_ll_b.next
    
    while cur_ll_a != cur_ll_b:
      cur_ll_a = cur_ll_a.next
      cur_ll_b = cur_ll_b.next
    
    return cur_ll_a
````
